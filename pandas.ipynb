## Creating a DataFrame from a List of Dictionaries

```python
dict = {'name': ["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score': [90, 40, 80, 98]}

df = pd.DataFrame(dict)
print(df)
```

## Python Pandas Series


```python
data = ["Apple", "Banana", "Cherry", "Date", "Elderberry" ]
s = pd.Series(data )
s.name = "Fruits"
print(s)
```
